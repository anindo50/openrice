```python
from bs4 import BeautifulSoup
import requests as req
from user_agents import parse
import pandas as pd
from multiprocessing import Pool

headers = {'User-Agent': 'parse'}

url = 'https://www.openrice.com/en/hongkong/restaurants'
```

```python
title = []
number = []
address = []
price_range = []
review = []
category = []
subcategory = []
bookmark = []
```

```python
while True:
    resp = req.get(url, headers = headers)
    data = resp.text
    soup = BeautifulSoup(data,'lxml')
    cons =  soup.find_all('section',class_='content-wrapper')
    for con in cons:
        titles = con.find(class_='title-name').text.replace('\n','')
        title.append(titles)
        link = con.find(class_='title-name')
        links = link.find('a').get('href')
        link = 'https://www.openrice.com' + links
        s_link = req.get(link, headers = headers)
        s_data = s_link.text
        s_soup = BeautifulSoup(s_data,'lxml')
        phone_c = s_soup.find(class_='telephone-section')
        if phone_c:
            phone = phone_c.find(class_='content').text
            print(phone)
            number.append(phone)
        else:
            'N/A'
        #for i in phone_c.find_all(class_='content'):
           # phone = i.text
            #print (phone)
        ad_c = con.find(class_='icon-info address')
        #addresses = ad_c.find(class_='content').text
        addresses = ad_c.span.text.replace('\n','').replace('\r','')
        address.append(addresses)
        r_c = con.find(class_='icon-info icon-info-food-price')
        price_ranges = r_c.span.text
        price_range.append(price_ranges)
        rv_c = con.find(class_='counters-container')
        reviews = rv_c.span.text
        review.append(reviews)
        sub_c = con.find(class_='pois-categoryui-list')
        categorys = sub_c.find('a').text
        category.append(categorys)
        subcategorys = sub_c.find('a').find_next('a').text
        subcategory.append(subcategorys)
        bookmarks = con.find('div',class_='text bookmarkedUserCount js-bookmark-count')['data-count']
        bookmark.append(bookmarks)
    url_tag = soup.find('a',class_='pagination-button next js-next')
    if url_tag:
        url = 'https://www.openrice.com' + url_tag.get('href')
    else:
        break
    #p = Pool(20)
    #p.map(soup,url)
    #p.terminate()
    #p.join()
```

```python
ti = pd.DataFrame({'Name':title})
ad = pd.DataFrame({'Address':address})
ph = pd.DataFrame({'Number':number})
re = pd.DataFrame({'Review':review})
bo = pd.DataFrame({'Bookmark':bookmark})
pr = pd.DataFrame({'Price_range':price_range})
ca = pd.DataFrame({'category':category})
sub = pd.DataFrame({'Subcategory':subcategory})
df = pd.concat([ti,ad,ph,re,bo,pr,ca,sub],axis = 1)
df.to_csv('openrice.csv',encoding = 'utf-8',header=True)
```
